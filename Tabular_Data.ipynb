###**Tabular_Data**

```python
from fastai.tabular.all import *
import pandas as pd

#dataset
path = untar_data(URLs.ADULT_SAMPLE)

#loading dataset into frame
df=pd.read_csv(path/'adult.csv')
df.head()
```

```python
df.info()
```

###**Train**

```python
#dataloaders
dls=TabularDataLoaders.from_csv(path/'adult.csv', path=path, bs=64, y_names='salary', skipinitialspace=False,
                                cat_names=['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex'],
                                cont_names= ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week'],
                                procs=[Categorify, FillMissing, Normalize])
#train
learn=tabular_learner(dls, metrics=accuracy)

#fit_one_cycle
learn.fit_one_cycle(4)
```

###**Test**

```python
df.iloc[[1,2]]
```

```python
#test
row0, pred0, probs0=learn.predict(df.iloc[1])
row1, pred1, probs1=learn.predict(df.iloc[2])

print(f'1: {pred0.item()} - {probs0[pred0.item()]*100: .1f}%')
print(f'2: {pred1.item()} - {probs1[pred1.item()]*100: .1f}%')
```

from the above result we can come the following conclusions:

*   1-row: meaning the model predicts class 1 (">=50K salary") with 73.5% confidence
*   2-row: meaning the model predicts class 0("<50K salary") with 99.2% confidence.
